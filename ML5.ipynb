Starting by importing tensorflow to create and trian the model
and numpy to manage the data

```
import tensorflow as tf
import numpy as np
```

supervised Machine Learning is all about figuring out an algorithm given a set of inputs and output. 

```
c = np.array([-40, -10, 0, 8, 15, 22, 38])
f = np.array([-40, 14, 32, 46, 59, 72, 100])
for i in range(len(c)-4):
  print(f"{c[i]} celsisus equals to {f[i]} farenheit")
print("...")
```

Creating the mdoel. It's an easy problem so we are using a single neuron model

units = 1 means that the model will have 1 neuron
input_shape = [1] means that the shape of the input will be big 1 item


```
l0 = tf.keras.layers.Dense(units=1, input_shape=[1])
```

```
model = tf.keras.Sequential([l0])
```

Compile our model

```
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
```

train our model "similar to sklearn"

```
history = model.fit(c,f, epochs=500, verbose=False)
print("Finish training the model")
```

```
print(model.predict([100]))
```
