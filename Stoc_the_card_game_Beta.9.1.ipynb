```python
"""
It is faster and when decks have finished they return corerctly. Still in Beta
"""
import random
from termcolor import colored
from IPython.display import clear_output


suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ("Ace",'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Jack', 'Queen', 'King')
values = {'Ace':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Jack':8, 'Queen':9, 'King':10}
#creating the deck of 40 cards
deck = []
for suit in suits:
    for rank in ranks:
        deck.append(f"{rank} of {suit}")
random.shuffle(deck)

pl1_cov_deck = deck[:20] #half deck to pl1
pl1_uncov_deck = []
cpu_cov_deck = deck[20:] #half to the cpu
cpu_uncov_deck = []
#decks in the middle of the table start empty
hearts = [] 
diamonds = []
spades = []
clubs = []
        
def card_manage_general(card): #function to decide where to put the drawn card (for middle table)
    if card == "Ace of Hearts" or len(hearts) != 0 and card.split(" ")[2] == "Hearts" and values[card.split(" ")[0]] == values[hearts[-1].split(" ")[0]]+1: 
        hearts.append(card)
        return True
    elif card == "Ace of Diamonds" or len(diamonds) != 0 and card.split(" ")[2] == "Diamonds" and values[card.split(" ")[0]] == values[diamonds[-1].split(" ")[0]]+1: 
        diamonds.append(card)
        return True
    elif card == "Ace of Spades" or len(spades) != 0 and card.split(" ")[2] == "Spades" and values[card.split(" ")[0]] == values[spades[-1].split(" ")[0]]+1: 
        spades.append(card)
        return True
    elif card == "Ace of Clubs" or len(clubs) != 0 and card.split(" ")[2] == "Clubs" and values[card.split(" ")[0]] == values[clubs[-1].split(" ")[0]]+1: 
        clubs.append(card)
        return True

#I have to do 2 other differnt function for each player    
def card_manage_pl1(card):
    executed = card_manage_general(card)
    if len(cpu_uncov_deck) != 0 and values[card.split(" ")[0]] == values[cpu_uncov_deck[-1].split(" ")[0]]+1 and executed != True:
        cpu_uncov_deck.append(card)
    elif executed != True:
        pl1_uncov_deck.append(card)
        return True
    
def card_manage_cpu(card):
    executed = card_manage_general(card)
    if len(pl1_uncov_deck) != 0 and values[card.split(" ")[0]] == values[pl1_uncov_deck[-1].split(" ")[0]]+1 and executed != True:
        pl1_uncov_deck.append(card)
    elif executed != True:
        cpu_uncov_deck.append(card)
        return True
        
def formatting(): #this functio gives a color to the cards and at the end there is the print statement with the right tags to give define the table
    if len(hearts) == 0:
        last_heart_card = colored("No card for HEARTS", "green")
    else:
        last_heart_card = colored(hearts[-1],"red")

    if len(clubs) == 0:
        last_club_card = colored("No card for CLUBS", "green")
    else:
        last_club_card = colored(clubs[-1],"blue")

    if len(spades) == 0:
        last_spade_card = colored("No card for SPADES", "green")
    else:
        last_spade_card = colored(spades[-1],"blue")

    if len(diamonds) == 0:
        last_diamond_card = colored("No card for DIAMONDS", "green")
    else:
        last_diamond_card = colored(diamonds[-1],"red")
        
    if len(cpu_uncov_deck) == 0:
        last_cpu_card = colored("First Turn","green")
    else:
        if cpu_uncov_deck[-1].split(" ")[-1] == "Spades" or cpu_uncov_deck[-1].split(" ")[-1] == "Clubs":
            last_cpu_card = colored(cpu_uncov_deck[-1],"blue")
        else:
            last_cpu_card = colored(cpu_uncov_deck[-1],"red")
    
    if len(pl1_uncov_deck) == 0:
        last_pl1_card = colored("First Turn","green")
    else:
        if pl1_uncov_deck[-1].split(" ")[-1] == "Spades" or pl1_uncov_deck[-1].split(" ")[-1] == "Clubs":
            last_pl1_card = colored(pl1_uncov_deck[-1],"blue")
        else:
            last_pl1_card = colored(pl1_uncov_deck[-1],"red")

    print(f"Table:\n\t\t\tCPU cards: {len(cpu_cov_deck)}\n\t\t\t{last_cpu_card}  cards: {len(cpu_uncov_deck)}\n\n\t{last_heart_card}\t\t{last_club_card}\n\t{last_spade_card}\t\t{last_diamond_card}\n\n\t\t\t{last_pl1_card}  cards: {len(pl1_uncov_deck)}\n\t\t\tYour cards: {len(pl1_cov_deck)}")
        

def pl1_plays(): #the function that contains all the action that pl1 must do
    global pl1_cov_deck,pl1_uncov_deck
    stop = False
    uncov_card = False
    while stop != True:
        while uncov_card != True:
            if len(pl1_uncov_deck) != 0:
                card = pl1_uncov_deck.pop()
                uncov_card = card_manage_pl1(card)
                if uncov_card == True:
                    break
                else:
                    print(colored(f"Player1 is playing with {card}: ","green"))
                    formatting()
                    input()
            else: 
                break

        if len(pl1_cov_deck) == 0 and len(pl1_uncov_deck) != 0:
            pl1_cov_deck = pl1_uncov_deck.copy()
            pl1_uncov_deck.clear()
        card = pl1_cov_deck.pop()
        stop = card_manage_pl1(card)
        print(colored(f"Player1 is playing with {card}: ","green"))
        formatting()
        input()
                  

def cpu_plays(): #the function that contains all the action that cpu must do
    global cpu_cov_deck,cpu_uncov_deck
    stop = False
    uncov_card = False
    while stop != True:
        while uncov_card != True:
            if len(cpu_uncov_deck) != 0:
                card = cpu_uncov_deck.pop()
                uncov_card = card_manage_cpu(card)
                if uncov_card == True:
                    break
                else:
                    print(colored(f"Nonno Carlo is playing with {card}: ","green"))
                    formatting()
                    input()
            else:
                break
        if len(cpu_cov_deck) == 0 and len(cpu_uncov_deck) != 0:
            cpu_cov_deck = cpu_uncov_deck.copy()
            cpu_uncov_deck.clear()
        card = cpu_cov_deck.pop()
        stop = card_manage_cpu(card)
        print(colored(f"Nonno Carlo is playing with {card}: ","green"))
        formatting()
        input()
    return True

def check_wins(): #function that checks if the game's ended and writes something if yes
    global pl1_cov_deck,pl1_uncov_deck,cpu_cov_deck,cpu_uncov_deck
    if len(pl1_uncov_deck) == 0 and len(pl1_cov_deck) == 0:
        print("\tCONGRATS,\nPLAYER1 WON")
    elif len(cpu_uncov_deck) == 0 and len(cpu_cov_deck) == 0:
        print("\tDID YOU REALLY LOSE VERSUS A STUPID PC?\n\tPC won,\t\nanyway")
        
def playing(): #function that runs the whole stuff
    while True:
        check_wins()
        pl1_plays()
        clear_output(wait = True)

        check_wins()
        cpu_plays()
        clear_output(wait = True)
playing()
```
