```python
#coin flip
"""
I'll use randint where 0 = Head and 1 = Tail

in for loop it asks for the tosses
    each time x = 1 or 0 
    if x = 1:
        print tail and the tails var counter += 1
    else (obv it's 0 so head):
        print head and hands counter += 1

print the results (how many tails, heads and the total of tosses)


the second is the same but if you wanna create suspence between each toss you have to press return 

"""
from random import randint 

def fast_coin_flip():
    tails = 0
    heads = 0
    
    for x in range(int(input("How many times: "))):
        x = randint(0,1)
        if x == 1:
            print("Tail!")
            tails += 1
        else:
            print("Head!")
            heads += 1
            
    print("\n Tails:", tails,"\n Head:", heads,"\n Total = ", tails + heads)
    
def slow_coin_flip():
    tails = 0
    heads = 0
    
    for x in range(int(input("How many times: "))):
        x = randint(0,1)
        if x == 1:
            print("Tail!")
            tails += 1
        else:
            print("Head!")
            heads += 1
        
        if input("Press ENTER to CONTINUE") == "":
            continue
     
    print("\n Tails:", tails,"\n Head:", heads,"\n Total = ", tails + heads)
    
def coin_flip():
    if input("Instant or Slowed?").lower()[0] == "i":
        fast_coin_flip()
    else:
        slow_coin_flip() 
```
